<h1>Задача 1. Дан набор из $p$ матриц размерностью $(n, n)$ и $p$ векторов размерностью $(n, 1)$, найти сумму произведений матриц на векторы. Написать тесты для кода

```python
def sum_prod(X, V):
    result = np.zeros((X[0].shape[0], 1))
    for i in range(len(X)):
        result += np.dot(X[i], V[i])
    return result

def test():
    result = np.array([[5], [11]])
    result1 = sum_prod([np.array([[1, 2], [3, 4]])], [np.array([[1], [2]])])
    assert np.array_equal(result1, result), 'Тест 1 не пройден'
    print('Все тесты пройдены')


test()
```

<h1>Задача 2. Дана матрица M, напишите функцию, которая бинаризует матрицу по некоторому threshold (то есть, все значения большие threshold становятся равными 1, иначе 0). Напишите тесты для кода

```python
def binarize(M, threshold=0.5):
    return np.where(M > threshold, 1, 0)


def test():
    result = binarize(np.array([[1, 3], [2, 4]]), 2)
    assert np.array_equal(result, np.array([[0, 1], [0, 1]])), 'Тест 1 не пройден'
    print('Все тесты пройдены')

test()
```

<h1>Задача 3. Напишите функцию, которая возвращает уникальные элементы из каждой строки матрицы. Напишите такую же функцию, но для столбцов. Напишите тесты для кода

```python
def unique_rows(mat):
    return [np.unique(row) for row in mat]


def unique_columns(mat):
    return [np.unique(mat[:, col]) for col in range(mat.shape[1])]

def test_rows():
    mat = np.array([[1, 2, 2],
                   [3, 3, 3],
                   [4, 5, 4]])

    result = unique_rows(mat)
    for i in range(len(result)):
        assert np.array_equal(result[i], [np.array([1, 2]), np.array([3]), np.array([4, 5])][i]), 'Тест 1 не пройден'
    print('Все тесты rows пройдены')

def test_columns():
    mat = np.array([[1, 2, 2],
                   [1, 3, 2],
                   [4, 3, 5]])

    result = unique_columns(mat)
    for i in range(len(result)):
        assert np.array_equal(result[i], [np.array([1, 4]), np.array([2, 3]), np.array([2, 5])][i]), 'Тест 2 не пройден'
    print('Все тесты columns пройдены')

test_rows()
test_columns()
```

<h1>Задача 4. Напишите функцию, которая заполняет матрицу с размерами $(m, n)$ случайными числами, распределенными по нормальному закону. Затем считает мат. ожидание и дисперсию для каждого из столбцов и строк, а также строит для каждой строки и столбца гистограмму значений (использовать функцию hist из модуля matplotlib.plot)

```python
def generate_matrix(m, n):
    mat = np.random.randn(m, n)

    row_means = np.mean(mat, axis=1)
    row_vars = np.var(mat, axis=1)

    col_means = np.mean(mat, axis=0)
    col_vars = np.var(mat, axis=0)

    print("Мат. ожидания по строкам:", row_means)
    print("Мат. ожидания по столбцам:", col_means)
    print("Дисперсии по строкам:", row_vars)
    print("Дисперсии по столбцам:", col_vars)

    for i in range(m):
        plt.figure()
        plt.hist(mat[i, :], bins=10, alpha=0.7, color='blue')
        plt.title(f'Гистограмма строки {i+1}')
        plt.xlabel('Значения')
        plt.ylabel('Частота')
        plt.grid(True)
        plt.show()

    for j in range(n):
        plt.figure()
        plt.hist(mat[:, j], bins=10, alpha=0.7, color='green')
        plt.title(f'Гистограмма столбца {j+1}')
        plt.xlabel('Значения')
        plt.ylabel('Частота')
        plt.grid(True)
        plt.show()


generate_matrix(2, 3)
```

<h1>Задача 5. Напишите функцию, которая заполняет матрицу $(m, n)$ в шахматном порядке заданными числами $a$ и $b$. Напишите тесты для кода

```python
def chess(m, n, a, b):
    result = np.zeros((m, n))

    for i in range(m):
        for j in range(n):
            if (i + j) % 2 == 0:
                result[i, j] = a
            else:
                result[i, j] = b
    return result


def test():
    assert np.array_equal(chess(3, 3, 1, 0), np.array([[1, 0, 1],
                                 [0, 1, 0],
                                 [1, 0, 1]])), 'Тест 1 не пройден'
    print('Все тесты пройдены')

test()
```

<h1>Задача 6. Напишите функцию, которая отрисовывает прямоугольник с заданными размерами (a, b) на изображении размера (m, n), цвет фона задайте в схеме RGB, как и цвет прямоугольника. Цвета также должны быть параметрами функции. Напишите аналогичную функцию но для овала с полуосями a и b. Напишите тесты для кода.
Примечание: уравнение эллипса (границы овала) можно записать как:
<h1>$\frac{(x-x_0)^2}{a^2}+\frac{(y-y_0)^2}{b^2}=1$

```python
def draw_rectangle(a, b, m, n, rectangle_color, background_color):
    fig, ax = plt.subplots(figsize=(m / 100, n / 100))
    ax.set_xlim(0, m)
    ax.set_ylim(0, n)
    ax.set_facecolor(background_color)
    rectangle = plt.Rectangle((0, 0), a, b, color=rectangle_color)
    ax.add_patch(rectangle)
    plt.axis('off')
    plt.show()


def draw_ellipse(a, b, m, n, ellipse_color, background_color):
    fig, ax = plt.subplots(figsize=(m / 100, n / 100))
    ax.set_xlim(0, m)
    ax.set_ylim(0, n)
    ax.set_facecolor(background_color)

    t = np.linspace(0, 2 * np.pi, 100)
    x = a * np.cos(t)
    y = b * np.sin(t)
    ax.fill(x + m / 2 - a / 2, y + n / 2 - b / 2, color=ellipse_color)

    plt.axis('off')
    plt.show()


# Тесты
def test_rectangle():
    draw_rectangle(100, 50, 200, 100, 'red', 'white')
    print("Тест прямоугольника завершен.")


def test_ellipse():
    draw_ellipse(100, 50, 200, 100, 'green', 'white')
    print("Тест овала завершен.")


test_rectangle()
test_ellipse()
```
